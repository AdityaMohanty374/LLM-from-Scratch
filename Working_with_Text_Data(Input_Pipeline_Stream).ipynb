```python
#simple tokenizer class
class SimpleTokenizerV1:
    def __init__(self, vocab):
      #mapping strings to integer in vocab
        self.str_to_int = vocab
      #mapping integers to strings in vocab
        self.int_to_str = {i:s for s,i in vocab.items()}

    def encode(self, text):
      #text to token
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
      #remove white spaces
        preprocessed = [
            item.strip() for item in preprocessed if item.strip()
        ]
      #assign the ids
        ids = [self.str_to_int[s] for s in preprocessed]
        return ids

    def decode(self, ids):
      #assign words from the ids
        text = " ".join([self.int_to_str[i] for i in ids])
        # Replace spaces before the specified punctuations
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
        return text
```

```python
#newer version of the simple tokenizer
class SimpleTokenizerV2:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = { i:s for s,i in vocab.items()}

    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]
        #if the item is present in the vocab then return its token if not return i
        preprocessed = [
            item if item in self.str_to_int
            else "<|unk|>" for item in preprocessed
        ]

        ids = [self.str_to_int[s] for s in preprocessed]
        return ids

    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        # Replace spaces before the specified punctuations
        text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
        return text
```

```python
import torch
from torch.utils.data import Dataset, DataLoader


class GPTDatasetV1(Dataset):
    def __init__(self, txt, tokenizer, max_length, stride):
        self.input_ids = []
        self.target_ids = []

        # Tokenize the entire text
        token_ids = tokenizer.encode(txt, allowed_special={"<|endoftext|>"})
        assert len(token_ids) > max_length, "Number of tokenized inputs must at least be equal to max_length+1"

        # Use a sliding window to chunk the book into overlapping sequences of max_length
        for i in range(0, len(token_ids) - max_length, stride):
            input_chunk = token_ids[i:i + max_length]
            target_chunk = token_ids[i + 1: i + max_length + 1]
            self.input_ids.append(torch.tensor(input_chunk))
            self.target_ids.append(torch.tensor(target_chunk))

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        return self.input_ids[idx], self.target_ids[idx]
```

```python
def create_dataloader_v1(txt, batch_size=4, max_length=256,
                         stride=128, shuffle=True, drop_last=True,
                         num_workers=0):

    # Initialize the tokenizer
    tokenizer = tiktoken.get_encoding("gpt2")

    # Create dataset
    dataset = GPTDatasetV1(txt, tokenizer, max_length, stride)

    # Create dataloader
    dataloader = DataLoader(
        dataset,
        batch_size=batch_size,
        shuffle=shuffle,
        drop_last=drop_last,
        num_workers=num_workers
    )

    return dataloader
```

# **Create TOKEN EMBEDDING**

```python
#Suppose we have the following four input examples with input ids 2, 3, 5, and 1 (after tokenization):
input_ids = torch.tensor([2, 3, 5, 1])
```

```python
#For the sake of simplicity, suppose we have a small
#vocabulary of only 6 words and we want to create embeddings of size 3,
#This would result in a 6x3 weight matrix
vocab_size = 50257  #not using 6 instead 50257 as its more pratical( BytePair encoder )
output_dim = 256   #not using 3 instead 256 as more pratical

torch.manual_seed(123)
embedding_layer = torch.nn.Embedding(vocab_size, output_dim)
print(embedding_layer.weight)
```

```python
#To convert a token with id 3 into a 3-dimensional vector, we do the following:
print(embedding_layer(torch.tensor([3])))
```

```python
#for all input ids
print(embedding_layer(input_ids))
```

# **Encoding word positions**

```python
max_length = 4
dataloader = create_dataloader_v1(
    raw_text, batch_size=8, max_length=max_length,
    stride=max_length, shuffle=False
)
data_iter = iter(dataloader)
inputs, targets = next(data_iter)
```

```python
print("Token IDs:\n", inputs)
print("\nInputs shape:\n", inputs.shape)
```

```python
token_embeddings = embedding_layer(inputs)
token_embeddings.shape
```

```python
token_embeddings[0, 0]
```

```python
#adding positional informations
#GPT-2 uses absolute position embeddings, so we just create another embedding layer, similar to the
#token embedding layer
context_length = max_length
pos_embedding_layer = torch.nn.Embedding(context_length, output_dim)
print(pos_embedding_layer.weight)
```

```python
pos_embeddings = pos_embedding_layer(torch.arange(max_length))
print(pos_embeddings.shape)
print(pos_embeddings)
```

```python
#To create the input embeddings used in an LLM
#we simply add the token and the positional embeddings
input_embeddings = token_embeddings + pos_embeddings
print(input_embeddings.shape)
print(input_embeddings)
```
